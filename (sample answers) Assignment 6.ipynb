Note: Two funcions differ by the level of indentation of `return False`. Show some example and demonstrate that these two functions can return different values.

````python
def func_A(my_list, target):
    for element in my_list:
        if element == target:
            return True
        return False

def func_B(my_list, target):
    for element in my_list:
        if element == target:
            return True
    return False
````

#### Q14.2 Create a complete version of `NYU_id` function. this function returns `True` for a sequence such as `'jk190'` that has 2-4 alphabets and 2-6 digits.

* The function should return `False` for invalid NetIDs, such as `'190jk'`, `'j190k'`.
* You may (and probably) use a variable `flag_alphabet`.

````python
alphabets = "abcdefghijklmnopqrstuvwxyz"
digits = "0123456789"

def NYU_id(astr):
    flag_alphabet = False
    flag_digit = False
    for char in astr:
        if char in alphabets:
            flag_alphabet = True
            if flag_digit:
                return False
        elif char in digits:
            flag_digit = True
            if not flag_alphabet:
                return False
        else:
            return False
    return flag_alphabet and flag_digit

result = NYU_id("jk190")
````

````python
result
````

The code above has two flags: Whether it receives an alphabet and a digit. 

For each iteration, 
* If one receives an alphabet, it should not have received a digit before that. Otherwise it returns False.
* If one receives a digit, it should have received an alphabet before. Otherwise it returns False.

After the iteration 
* It should have received alphabet(s) and digit(s).

This works with any alphabet(s) + digit(s). Limitation is that it did not count the number of alphabets and digits, we need more details if we want to control the length of them.
